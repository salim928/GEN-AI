```python
# Low level understanding of the tokenizer

from transformers import AutoTokenizer

#Load a pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# example of text
text = "I was so not happy with the next movie"

# tokenize text
tokens = tokenizer.tokenize(text)

# print tokens
print("Tokens:",tokens)

```

```python
# convert tokens into input id's
input_ids = tokenizer.convert_tokens_to_ids(tokens)

# print input ids
print("INPUT ID'S:",input_ids)

```

```python
# encode text (tokenization +  converting to input id's)
input_ids = tokenizer.encode(text)

# print input ids
print("Encoded Id's:",input_ids)

```

```python
# decode the text
decoded_text = tokenizer.decode(input_ids)

# print decoded text
print("Decoded text:",decoded_text)
```

# Fine-tuning IMDB

### Step 1: installing necessary **libraries**

```python
!pip install datasets
```

### Setp 2. Load  and prepare the dataset

```python
from datasets import load_dataset
dataset = load_dataset("imdb")
```

```python
dataset
```

```python
dataset["train"][0]
```

### Step 3.  Preprocess the Data

*Tokenize* the dataset using the tokenizer assocaited with the pre-trained **model**

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

#  Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

#mapping the dataset with the function to iterate and perform it's function
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

```python
tokenized_datasets['train'][0]
```

### Step 4. Set Up the Training Arguments

Specify the hyperparameters and training settings

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # output directory
    eval_strategy="epoch",           # evaluate every epoch
    learning_rate=2e-5,               # learning rate
    num_train_epochs=1,              # total number of training epochs
    per_device_train_batch_size=16,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
   # warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
)
training_args
```

### Sstep 5. Initialize the Model

Load the pre-trained model and define the training procedure

```python
from transformers import AutoModelForSequenceClassification, Trainer


# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# intialize the Trainer

trainer = Trainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=tokenized_datasets["train"],         # training dataset
    eval_dataset=tokenized_datasets['test']
)
```

### Step 6. Train the Model

Fine-tune the pre-trained model on your specific dataset



```python
import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
```

```python
# Train the model
import torch
with torch.autocast(device_type='cuda', dtype=torch.float16):
    trainer.train()
```

### Step 7. Evaluate the Model
Save the fine-tuned model for later use.

```python
# Save the Model

model.save_pretrained("./results")
tokenizer.save_pretrained("./fine-tuned-model")
```

```python
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer

# Load the fine-tuned model and tokenizer
model = AutoModelForSequenceClassification.from_pretrained("./results")
tokenizer = AutoTokenizer.from_pretrained("./fine-tuned-model")
```

```python
# Evaluate
results = trainer.evaluate()
print(results)
```

```python
import arxiv
import pandas as pd
```

```python
# Query to fetch AI-related paper
query = "ai Or artificial intelligence Or machine learning"
search = arxiv.Search(
    query=query,
    max_results=10,
    sort_by=arxiv.SortCriterion.LastUpdatedDate
)

# Fetch papers
papers = []
for result in search.results():
    papers.append({
        "title": result.title,
        'abstract': result.summary,
        "authors": result.authors,
        "published": result.published,
        "category": result.categories,
    })

# convert to DataFrame
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
# Research paper summarization system
# Example abstract from api
abstract = df['abstract'][0]

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarization
summarization = summarizer(abstract)
```

```python
summarization_result = summarization[0]['summary_text']
print(summarization_result)
```
